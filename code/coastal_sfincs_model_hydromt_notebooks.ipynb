```python
sfincs_root_norain = "./xaver_cuxhaven_norain_sfincs_GTSM/"  # (relative) path to sfincs root
sfincs_root_rain = "./xaver_cuxhaven_rain_sfincs_GTSM/"  # (relative) path to sfincs root
mod_nr = SfincsModel(sfincs_root_norain, mode="r")
mod_r = SfincsModel(sfincs_root_rain, mode="r")

# we can simply read the model results (sfincs_map.nc and sfincs_his.nc) using the read_results method
mod_nr.read_results()
mod_r.read_results()

# the following variables have been found
list(mod_nr.results.keys())
```

```python
fig, ax = mod_nr.plot_basemap(fn_out=None, bmap="sat", figsize=(11, 7))
plt.show()
fig, ax = mod_r.plot_basemap(fn_out=None, bmap="sat", figsize=(11, 7))
plt.show()
```

```python
# write hmax to <mod.root>/gis/hmax.tif
mod_nr.write_raster("results.hmax", compress="LZW")
mod_r.write_raster("results.hmax", compress="LZW")
```

```python
# read global surface water occurance (GSWO) data to mask permanent water
# NOTE to read data for a different region than Northen Italy add this data to the data catalog:
mod_nr.data_catalog.from_yml('deltares_data.yml')
gswo = mod_nr.data_catalog.get_rasterdataset("gswo", buffer=10)
# permanent water where water occurence > 5%
gswo_mask = gswo.raster.reproject_like(mod_nr.staticmaps, method="max") <= 5
```

```python
hmin = 0.2  # minimum flood depth [m] to plot
da_hmax = mod_nr.results["hmax"]  # hmax is computed from zsmax - zb
# get overland flood depth with GSWO and set minimum flood depth
da_hmax_fld = da_hmax.where(gswo_mask).where(da_hmax > hmin)
# update attributes for colorbar label later
da_hmax.attrs.update(long_name="flood depth", unit="m")
```

```python
(mod_nr.results['hmax'] - mod_r.results['hmax'] ).plot() 
```

```python
h = mod_nr.results["h"].where(mod_nr.results["h"] > 0)
h.attrs.update(long_name="water depth", unit="m")
h.sel(time=h["time"].values[4::4]).plot(col="time", col_wrap=3, vmax=4)
```

```python
# point_h contains the water depths at the sfincs.obs gauge locations
# see mod.plot_basemaps (or next figure) for the location of the observation points
mod_nr.data_catalog.from_yml('./deltares_data.yml')
gswo = mod_nr.data_catalog.get_rasterdataset("gswo", buffer=10)
# permanent water where water occurence > 5%
gswo_mask = gswo.raster.reproject_like(mod_nr.staticmaps, method="max") <= 5
```

```python
hmin = 0.2  # minimum flood depth [m] to plot
da_hmax = mod_nr.results["hmax"]  # hmax is computed from zsmax - zb
# get overland flood depth with GSWO and set minimum flood depth
da_hmax_fld = da_hmax.where(gswo_mask).where(da_hmax > hmin)
# update attributes for colorbar label later
da_hmax.attrs.update(long_name="flood depth", unit="m")
```

```python
# create hmax plot and save to mod.root/figs/hmax.png
fig, ax = mod_nr.plot_basemap(
    fn_out=None,
    variable=None,
    bmap="sat",
    geoms=["src", "obs"],
    plot_bounds=False,
    figsize=(11, 11),
)
# plot overland flooding based on gswo mask and mimum flood depth
cbar_kwargs = {"shrink": 0.6, "anchor": (0, 0)}
cax_fld = da_hmax_fld.plot(
    ax=ax, vmin=0, vmax=3.0, cmap=plt.cm.viridis, cbar_kwargs=cbar_kwargs
)

ax.set_title(f"SFINCS maximum water depth")
plt.savefig(join(mod_nr.root, '', 'hmax.png'), dpi=225, bbox_inches="tight")
```

```python
# create hmax plot and save to mod.root/figs/hmax.png
fig, ax = mod_r.plot_basemap(
    fn_out=None,
    variable=None,
    bmap="sat",
    geoms=["src", "obs"],
    plot_bounds=False,
    figsize=(11, 11),
)
# plot overland flooding based on gswo mask and mimum flood depth
cbar_kwargs = {"shrink": 0.6, "anchor": (0, 0)}
cax_fld = da_hmax_fld.plot(
    ax=ax, vmin=0, vmax=3.0, cmap=plt.cm.viridis, cbar_kwargs=cbar_kwargs
)

ax.set_title(f"SFINCS maximum water depth")
plt.savefig(join(mod_r.root, '', 'hmax.png'), dpi=225, bbox_inches="tight")
```
